```python
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

import numpy as np

from PIL import Image
from skimage import transform

from keras.models import load_model
```

```python
def load_image(filename):
    img = Image.open(filename)
    np_img = np.array(img).astype('float32') / 255
    np_img = transform.resize(np_img, (224, 224, 3))
    np_img = np.expand_dims(np_img, axis=0)

    img = mpimg.imread(filename)
    plt.imshow(img, cmap='gray')

    return np_img
```

```python
MODEL_NAME = 'model_saved_100epochs.h5'
FILENAME = 'dataset/test/2023-06-08_165712-871277.jpg'
```

```python
# Load Model CNN
print(MODEL_NAME)
model = load_model('model/' + MODEL_NAME)
print(model)
# Load Image
image = load_image(FILENAME)
```

```python
label = model.predict(image)[0]

label = list(label)

max_index = label.index(max(label))
max_index
```
